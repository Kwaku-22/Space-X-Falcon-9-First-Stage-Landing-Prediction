```python
## Task 1: Mark all launch sites on a map

# Define the launch site locations
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Define the coordinates of the closest coastline (example: coastline near CCAFS LC-40)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Create the map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=14)

# Define the coordinates for the PolyLine (launch site and coastline)
coordinates = [
    [launch_site_lat, launch_site_lon],  # Launch site coordinates
    [coastline_lat, coastline_lon]       # Coastline coordinates
]

# Create a `folium.PolyLine` object
line = folium.PolyLine(locations=coordinates, weight=2, color='blue')

# Add the PolyLine to the map
site_map.add_child(line) 

# Add markers for the launch site and coastline
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='Launch Site (CCAFS LC-40)',
    icon=folium.Icon(color='green')
).add_to(site_map)

folium.Marker(
    location=[coastline_lat, coastline_lon],
    popup='Coastline',
    icon=folium.Icon(color='red')
).add_to(site_map)

# Display the map
site_map.save("launch_map.html")  # Saves the map as an HTML file

site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Define the Haversine function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of Earth in kilometers
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Existing launch sites DataFrame
launch_sites_df = pd.DataFrame({
    "Launch Site": ["CCAFS LC-40", "CCAFS SLC-40", "KSC LC-39A", "VAFB SLC-4E"],
    "Lat": [28.562302, 28.563197, 28.573255, 34.632834],
    "Long": [-80.577356, -80.576820, -80.646895, -120.610745]
})

# Define proximities (e.g., coastline, city center)
proximities = [
    {"name": "Coastline", "lat": 28.56367, "long": -80.57163},  # Example coastline
    {"name": "City Center", "lat": 28.608389, "long": -80.807223}  # Example city center
]

# Compute distances between launch sites and proximities
distances = []
for index, site in launch_sites_df.iterrows():
    site_lat = site['Lat']
    site_lon = site['Long']
    site_name = site['Launch Site']
    
    for proximity in proximities:
        proximity_name = proximity["name"]
        proximity_lat = proximity["lat"]
        proximity_lon = proximity["long"]
        distance_km = haversine(site_lat, site_lon, proximity_lat, proximity_lon)
        distances.append({
            "Launch Site": site_name,
            "Proximity": proximity_name,
            "Distance (km)": round(distance_km, 2)
        })

# Convert to DataFrame
distances_df = pd.DataFrame(distances)

# Display the results
print(distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from math import radians, sin, cos, sqrt, atan2

# Define the Haversine function to calculate distances
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of Earth in kilometers
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Define the launch site coordinates
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Define the coordinates of a point of interest (e.g., city, railway, highway)
poi_lat = 28.401570  # Example latitude (replace with actual value)
poi_lon = -80.605660  # Example longitude (replace with actual value)
poi_name = "Closest City"  # Example name

# Calculate the distance between the launch site and the point of interest
distance_km = haversine(launch_site_lat, launch_site_lon, poi_lat, poi_lon)

# Initialize the map centered at the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add a marker for the launch site
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='Launch Site (CCAFS LC-40)',
    icon=folium.Icon(color='green')
).add_to(site_map)

# Add a marker for the point of interest
folium.Marker(
    location=[poi_lat, poi_lon],
    popup=f"{poi_name}<br>Distance: {round(distance_km, 2)} km",
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Draw a PolyLine between the launch site and the point of interest
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [poi_lat, poi_lon]],
    color='blue',
    weight=2.5
).add_to(site_map)

# Save and display the map
site_map.save("launch_site_to_poi.html")
site_map
```
